```python
df = pd.read_csv('Unemployment in India.csv')

df.columns = df.columns.str.strip()

df.head()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df['Estimated Unemployment Rate (%)'],
    bins=20,
    kde=True
)

plt.title("Distribution of Unemployment Rate")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.barplot(
    x='Region',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.xticks(rotation=90)

plt.title("Unemployment Rate by Region")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='Area',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title("Urban vs Rural Unemployment")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
plt.figure(figsize=(14,6))

sns.lineplot(
    x='Date',
    y='Estimated Unemployment Rate (%)',
    data=df
)

plt.title("Unemployment Trend Over Time")

plt.show()
```

```python
top_regions = df.groupby(
    'Region'
)['Estimated Unemployment Rate (%)'].mean()

top_regions = top_regions.sort_values(
    ascending=False
)

print(top_regions.head(10))
```

```python
region_data = df.groupby(
    'Region'
)['Estimated Unemployment Rate (%)'].mean()

plt.figure(figsize=(8,8))

region_data.head(10).plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.ylabel("")

plt.title("Top Regions by Unemployment")

plt.show()
```

```python
# Final Insights

print("Dataset Shape:", df.shape)
print("\nNumber of Regions:", df['Region'].nunique())

print("\nAverage Unemployment Rate:")
print(round(df['Estimated Unemployment Rate (%)'].mean(), 2))

print("\nHighest Unemployment Region:")
highest = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().idxmax()
print(highest)

print("\nLowest Unemployment Region:")
lowest = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean().idxmin()
print(lowest)
```
